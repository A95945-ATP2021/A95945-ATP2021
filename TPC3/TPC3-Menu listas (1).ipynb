```python
def menu ():
    print ("""Gestão de listas de inteiros
              (1) Introduza a lista à mão;
              (2) Gerar uma lista com valores aleatórios;
              (3) Maior elemento da lista
              (4) Ordenar lista
              (5) Inverter ordem da lista
              (0) Sair""")
def criarListaM():
    n=int(input("Introduza o número de elementos da lista: "))
    L=[]
    for i in range (n):
        elem=int(input("Introduz o elemento"+ str(i+1)+":"))
        L.append(elem)
    return L

def criarListaA():
    import random
    n=int(input("Introduza o número de elementos da lista: "))
    L=[]
    for i in range (n):
        elem=random.randint (0,100)
        L.append(elem)
    return L

def maiorElemento(lista):
    maior=lista[0]
    i=1
    while (i<= len(lista)-1):
        if lista[i]>maior:
            maior=lista[i]
        i=i+1
    return maior

def ordenar(lista):
    inicio=1
    comparar=0
    while (comparar<len(lista)):
        menor=lista[comparar]
        inicio=comparar+1
        while inicio<len(lista):
            if menor>lista[inicio]:
                z=lista[inicio]
                lista[inicio]=menor
                lista[comparar]=z
                menor=z
            inicio+=1
        comparar+=1
    return lista

def inverterOrdem (lista):
    i=0
    x=len(lista)-1
    while i!=((x)/2):
        z=lista[i]
        w=lista[x-i]
        lista[i]=w
        lista[x-i]=z
        i+=1
    return lista

op="1"
while (op!="0"):
    menu()
    op=input("Qual é a sua opção:")
    if (op=="1"):
        lista=criarListaM()
        print (lista)
    elif (op=="2"):
        lista=criarListaA()
        print (lista)
    elif (op=="3"):
        maior=maiorElemento(lista)
        print (maior)
    elif (op=="4"):
        olista=ordenar(lista)
        print (olista)
    elif (op=="5"):
        plista=inverterOrdem(lista)
        print (plista)
    elif (op=="0"):
        print ("Até à próxima!")
    else: 
        print ("Opção inválida!")
```
