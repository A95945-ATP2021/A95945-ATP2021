```python
def menu ():
    print ("""Operações com frações:
              (1) Criar uma fração;
              (2) Simplificar uma fração;
              (3) Somar;
              (4) Subtrair;
              (5) Multiplicar;
              (6) Dividir
              (7) Gerar lista com n frações;
              (8) Somar lista;
              (9) Maior fração;
              (10)Maior Fração de uma lista;
              (11)Menor Fração de uma lista;
              (0) Sair""")


def criarFracao (numerador, denominador):
    return (numerador, denominador)

def verFracao (f):
    print (str (f[0])+("/")+ str (f[1]))

def mdc(a,b):
    while b !=0:
        resto = a % b
        a = b
        b = resto

    return a

def simplificarFracao(f):
    num,denom=f
    a=mdc(num,denom)
    return (num/a,denom/a)

def somaFracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return verFracao( simplificarFracao(criarFracao((n1*d1+n2*d2),d1*d2)))

def subtrairFracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return verFracao(simplificarFracao(criarFracao((n1*d1-n2*d2),d1*d2)))

def multiplicarFracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return verFracao(simplificarFracao(criarFracao((n1*n2),(d1*d2))))

def dividirFracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    return verFracao(simplificarFracao(criarFracao((n1*d2),(d1*n2))))

def gerarLista():
    import random
    listaFrac=[]
    n=int(input("Introduza o número de elementos da lista: "))
    for i in range (n):
        nume=random.randint (0,100)
        denomi=random.randint (1,100)
        f=criarFracao(nume,denomi)
        listaFrac.append (f)
    return listaFrac
    
    
def somarLista(listaFrac):
    ac=(0,1) #necessario variavel acumulativa 
    for elem in listaFrac:
        ac=somaFracao(ac,elem)
    return ac

def maiorFracao (f1,f2):
    n1,d1=f1
    n2,d2=f2
    n3=n1*d2
    n4=n2*d1
    if n3>n4:
        return f1
    else:
        return f2

def menorFracao(f1,f2):
    n1,d1=f1
    n2,d2=f2
    n3=n1*d2
    n4=n2*d1
    if n3<n4:
        return f1
    else:
        return f2

def maiorFracaoLista(listaFrac):
    maior=(0,0)
    for elem in listaFrac:
        maior=maiorFracao(maior,elem)
    return maior
        
def menorFracaoLista(listaFrac):
    menor=listaFrac[0]
    for elem in listaFrac:
        menor=menorFracao(menor,elem)
    return menor

op="1"
while (op!="0"):
    menu()
    op=input("Qual é a sua opção:")
    if (op=="1"):
        a=int(input("Numerador: "))
        b=int(input("Denominador: "))
        f1=simplificarFracao(criarFracao(a,b))
        verFracao(f1)
    elif (op=="2"):
        fracao=simplificarFracao(f1)
        verFracao (fracao)
   
    elif (op=="3"):
        print ("Introduza o numerador e denominador para a primeira fração.")
        a=int(input("Numerador: "))
        b=int(input("Denominador: "))
        f1=simplificarFracao(criarFracao(a,b))
        print ("Introduza o numerador e denominador para a segunda fração.")
        c=int(input("Numerador: "))
        d=int(input("Denominador: "))
        f2=simplificarFracao(criarFracao(c,d))
        soma=somaFracao (f1,f2)
        print (soma)
    
    elif (op=="4"):
        print ("Introduza o numerador e denominador para a primeira fração.")
        a=int(input("Numerador: "))
        b=int(input("Denominador: "))
        f1=simplificarFracao(criarFracao(a,b))
        print ("Introduza o numerador e denominador para a segunda fração.")
        c=int(input("Numerador: "))
        d=int(input("Denominador: "))
        f2=simplificarFracao(criarFracao(c,d))
        subtrair=subtrairFracao(f1,f2)
        print (subtrair)
    
    elif (op=="5"):
        print ("Introduza o numerador e denominador para a primeira fração.")
        a=int(input("Numerador: "))
        b=int(input("Denominador: "))
        f1=simplificarFracao(criarFracao(a,b))
        print ("Introduza o numerador e denominador para a segunda fração.")
        c=int(input("Numerador: "))
        d=int(input("Denominador: "))
        f2=simplificarFracao(criarFracao(c,d))
        multiplicacao=multiplicarFracao (f1,f2)
        print (multiplicacao)
    
    elif (op=="6"):
        print ("Introduza o numerador e denominador para a primeira fração.")
        a=int(input("Numerador: "))
        b=int(input("Denominador: "))
        f1=simplificarFracao(criarFracao(a,b))
        print ("Introduza o numerador e denominador para a segunda fração.")
        c=int(input("Numerador: "))
        d=int(input("Denominador: "))
        f2=simplificarFracao(criarFracao(c,d))
        divisao=dividirFracao (f1,f2)
        print (divisao)
        
    elif (op=="7"):
        l=gerarLista()
        print (l)
    
    elif (op=="8"):
        s=somarLista(l)
        print (s)
    elif (op=="9"):
        print ("Introduza o numerador e denominador para a primeira fração.")
        a=int(input("Numerador: "))
        b=int(input("Denominador: "))
        f1=(criarFracao(a,b))
        print ("Introduza o numerador e denominador para a segunda fração.")
        c=int(input("Numerador: "))
        d=int(input("Denominador: "))
        f2=(criarFracao(c,d))
        d=maiorFracao(f1,f2)
        print(d)
        
    elif (op=="10"):
        r=maiorFracaoLista(l)
        print(r)
    elif (op=="11"):
        p=menorFracaoLista(l)
        print(p)
    elif (op=="0"):
        print ("Até à próxima!")
    else: 
        print ("Opção inválida!")
```
